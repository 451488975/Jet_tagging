```python
import pandas as pd
import h5py
from general_transform import Transform
```

```python
df = Transform('v0/train.h5', 'v0', 'h5')
```

```python
df
```

```python
features = df.columns.values
features = features[2:]
features
```

```python
import matplotlib.pyplot as plt
def plot1D(fea):
    val = df[fea].values

    t = []

    for x in val:
        t.append(x)
    plt.hist(t, 100, histtype='step', density=True, log=True, label='top', color='purple')
    plt.xlabel(fea)
    plt.ylabel('Prob Density')
    plt.legend()
    plt.show()
```

```python
for fea in features:
    plot1D(fea)
```
